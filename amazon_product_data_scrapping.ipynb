```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://www.amazon.in/s?k=ps5+games&crid=1S0WX986VWMPY&sprefix=ps5+game%2Caps%2C244&ref=nb_sb_noss_1'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36 OPR/119.0.0.0',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
webpage = requests.get(url, headers=headers)
```

```python
webpage
```

```python
webpage.content
```

```python
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
soup.prettify()
```

```python
soup.find_all('span')
```

```python
product = soup.find_all('a', attrs={'class':'a-link-normal'})
```

```python
product[0]
```

```python
product1 = product[0].get('href')
```

```python
product_link = 'https://amazon.in' + product1
```

```python
product_link
```

Now moving to game details page

```python
new_webpage = requests.get(product_link, headers=headers)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content,'html.parser')
```

```python
new_soup
```

```python
product_title=new_soup.find('span',id = 'productTitle').text.strip()
product_title
```

```python
product_price = new_soup.find('span',class_ = 'a-price-whole').text
product_price
```

```python
product_rating = new_soup.find('i',class_='a-icon a-icon-star a-star-4 cm-cr-review-stars-spacing-big').text
product_rating
```

```python
product_age_group = new_soup.find('a',class_='a-popover-trigger a-declarative').text.strip()
product_age_group
```
