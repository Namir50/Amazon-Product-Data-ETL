```python
import pandas as pd
import numpy as np
```

```python
amazon_df = pd.read_csv('amazon_product_data')
```

```python
amazon_df
```

```python
amazon_df.shape
```

```python
amazon_df['title'] = amazon_df['title'].replace("",np.nan)
amazon_df['price'] = amazon_df['price'].replace("",np.nan)
amazon_df['rating'] = amazon_df['rating'].replace("",np.nan)
amazon_df['no_of_ratings'] = amazon_df['no_of_ratings'].replace("",np.nan)
```

```python
amazon_df.isnull().sum()
```

```python
amazon_df = amazon_df.dropna(subset=['title'])
```

```python
amazon_df.isnull().sum()
```

```python
amazon_df.shape
```

```python
amazon_df
```

```python
amazon_df = amazon_df.drop_duplicates(subset='title').reset_index(drop=True)
```

```python
amazon_df
```

```python
amazon_df['price'] = amazon_df['price'].str.replace(',', '', regex=False).astype(int)
```

```python
amazon_df
```

```python
amazon_df['rating'] = amazon_df['rating'].str.split('out').str[0]
```

```python
amazon_df
```

```python
amazon_df.rename(columns={'title':'PS5 Game'},inplace=True)
```

```python
amazon_df
```

```python
amazon_df['rating'] = pd.to_numeric(amazon_df['rating'],errors='coerce')
```

```python
amazon_df['rating'] = amazon_df['rating'].fillna(amazon_df['rating'].mean())
```

```python
amazon_df['rating'] = amazon_df['rating'].round(2)
```

```python
amazon_df
```

```python
amazon_df['no_of_ratings'] = amazon_df['no_of_ratings'].str.replace('ratings','',case=False,regex=True)
```

```python
amazon_df
```

```python
amazon_df['no_of_ratings'] = amazon_df['no_of_ratings'].str.replace(',','',case=False,regex=False)
```

```python
amazon_df
```

```python
amazon_df['no_of_ratings'] = pd.to_numeric(amazon_df['no_of_ratings'],errors='coerce')
```

```python
amazon_df
```

```python
amazon_df['total_score'] = amazon_df['rating'] * amazon_df['no_of_ratings']
```

```python
amazon_df
```

```python
amazon_df.to_csv('amazon_product_data',index=False)
```
